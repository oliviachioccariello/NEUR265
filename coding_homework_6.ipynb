```python
import matplotlib.pyplot as plt   #import matplot
import numpy as np #import numpy 
```

```python
url1 = 'https://raw.githubusercontent.com/oliviachioccariello/NEUR265/main/acc_hits_hw6.csv' #import acc_hits file
acc_hits = np.loadtxt(url1, delimiter = ',') #create data variable

url2 = 'https://raw.githubusercontent.com/oliviachioccariello/NEUR265/main/acc_fa_hw6.csv' #import acc_fa file
acc_fa = np.loadtxt(url2, delimiter = ',') #create data variable

url3 = 'https://raw.githubusercontent.com/oliviachioccariello/NEUR265/main/lc_hits_hw6.csv' #import lc_hits file
lc_hits = np.loadtxt(url3, delimiter = ',') #create data variable

url4 = 'https://raw.githubusercontent.com/oliviachioccariello/NEUR265/main/lc_fa_hw6.csv' #import lc_fa file
lc_fa = np.loadtxt(url4, delimiter = ',') #create data variable
```

```python
dt = time[1] - time[0]       # Define the sampling interval.
K = 8001          # Define the number of trials.
N = acc_hits.shape[1]
# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  # ... for each trial in Acc_hits
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  # ... and each trial in lc_hits

# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of acc_hits trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and lc_hits trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.
plt.plot(f, cohr.real)                      # Plot coherence vs frequency,
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between Acc_hits and Lc_hits')
```

```python
dt = time[1] - time[0]       # Define the sampling interval.
K = 8001          # Define the number of trials.
N = acc_hits.shape[1]
# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_hits])  # ... for each trial in Acc_hits
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_hits])  # ... and each trial in lc_hits

# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of acc_hits trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and lc_hits trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.
plt.plot(f, cohr.real, color = 'red')                      # Plot coherence vs frequency,
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')

# Compute the Fourier transforms
xf = np.array([np.fft.rfft(x - x.mean()) for x in acc_fa])  # ... for each trial in Acc_fa
yf = np.array([np.fft.rfft(y - y.mean()) for y in lc_fa])  # ... and each trial in La_fa

# Compute the spectra
T = time[-1]
scale = 2 * dt**2 / T
Sxx = scale * (xf * xf.conj()).mean(0)  # Spectrum of Acc_fa trials
Syy = scale * (yf * yf.conj()).mean(0)  # ... and La_Fc trials
Sxy = scale * (xf * yf.conj()).mean(0)  # ... and the cross spectrum

# Compute the coherence.
cohr = np.abs(Sxy) / (np.sqrt(Sxx) * np.sqrt(Syy))

f = np.fft.rfftfreq(N, dt)                     # Define a frequency axis.
plt.plot(f, cohr.real, color = 'blue')                      # Plot coherence vs frequency,
plt.xlim([0, 50])                           # ... in a chosen frequency range,
plt.ylim([0, 1])                            # ... with y-axis scaled,
plt.xlabel('Frequency [Hz]')                # ... and with axes labeled.
plt.ylabel('Coherence')
plt.title('Coherence between Hits and False Alarms')
```
