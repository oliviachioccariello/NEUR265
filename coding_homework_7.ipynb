13. In a new *Code Cell*, calculate the firing rate across time for each of your four arrays (trial 1 left-turn, trial 1 right-turn, trial 2 left-turn, and trial 2 right-turn). Set your time bins equal to start at 0 s, end at 30 s (the length of the waiting period), with a bin size equal to 1 s. Create two subplots, and plot the firing rate over time for each trial in each of the two subplots. For example, your first subplot should have the firing rate across time for trial 1 left-turn, and trial 1 right-turn, and your second subplot should have the firing rate across time for trial 2 left-turn, and trial 2 right-turn. Make the line for right-turn trials blue, and the line for left-turn trials red. Label your axes.


```python
binsize = 1                                                                   # Make 1-second bins
fr_bins = np.arange(0,30,binsize)                                             # Make time bins (in 1 second increments) from 0 to 30 s

fr_left1 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array                                 
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index1 = (spikes_left_trial1 > fr_bins[i]) & (spikes_left_trial1 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_left_trial1[spk_index1]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_left1[i] = (len(true_values))/binsize    

fr_right1 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array                                 
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index2 = (spikes_right_trial1 > fr_bins[i]) & (spikes_right_trial1 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_right_trial1[spk_index2]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_right1[i] = (len(true_values))/binsize   

fr_left2 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array                                 
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index3 = (spikes_left_trial2 > fr_bins[i]) & (spikes_left_trial2 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_left_trial2[spk_index3]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_left2[i] = (len(true_values))/binsize    

fr_right2 = np.zeros(len(fr_bins)-1)                                             # Initialize our firing rate array                                 
for i in range(len(fr_bins)-1):                                               # Loop through each bin
  spk_index4 = (spikes_right_trial2 > fr_bins[i]) & (spikes_right_trial2 < fr_bins[i+1])           # Find timestamp values falling within each bin
  true_values = spikes_right_trial2[spk_index4]                                          # Our index is a boolean, so we need to isolate "True" values
  fr_right2[i] = (len(true_values))/binsize   
```

```python
fig, axes = plt.subplots(1, 2,figsize=(12, 3))

axes[0].plot(fr_bins[:-1], fr_left1, color = 'orange')
axes[0].plot(fr_bins[:-1], fr_right1)
axes[1].plot(fr_bins[:-1], fr_left2, color = 'orange')
axes[1].plot(fr_bins[:-1], fr_right2)
```
