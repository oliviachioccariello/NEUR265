```python
#make predictor arrays
shock_array = np.zeros(10458)
for i in range(10458):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(10458)
for i in range(10458):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0
```

```python
from pandas import DataFrame as df
import statsmodels.api as sm
from statsmodels.genmod.families import Poisson
from statsmodels.genmod.families.links import identity, log
from scipy.stats import chi2
from statsmodels.distributions.empirical_distribution import ECDF

first_trace = conditioning_traces [:, 0]
predictors_conditioning = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array}) #make predictors_conditioning 
predictors_conditioning['shock onset'] = shock_array
modelA = sm.GLM(first_trace, predictors_conditioning, family=Poisson())
modelA_results = modelA.fit() # Fit model to our data
```

```python
#create predictors variable for conditioning  

first_trace = conditioning_traces [:, 0]
predictors_conditioning['shock onset'] = shock_array
r_model_conditioning = np.zeros(63)

for i in range (63):
  trace_temp = conditioning_traces[:, i]
  modelA = sm.GLM(first_trace, predictors_conditioning, family=Poisson())
  modelA_results = modelA.fit() # Fit model to our data
  predicted_trace = modelA_results.predict()
  rA = pearsonr(trace_temp, predicted_trace)
  r_model_conditioning[i] = rA.statistic
```

```python
#Create a loop for extinction_traces 

second_trace = extinction_traces[:, 0] #make second_trace variable 

shock_array = np.zeros(14208)
for i in range(14208):
  if i >= (shock_index[0]-15)*15 and i <= (shock_index[0]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[1]-15)*15 and i <= (shock_index[1]+30)*15:
    shock_array[i] = 1
  elif i>= (shock_index[2]-15)*15 and i <= (shock_index[2]+30)*15:
    shock_array[i] = 1
  else:
    shock_array[i] = 0

tone_array = np.zeros(14208)
for i in range(14208):
  if i >= (tone_index[0]-15)*15 and i <= (tone_index[0]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[1]-15)*15 and i <= (tone_index[1]+30)*15:
    tone_array[i] = 1
  elif i>= (tone_index[2]-15)*15 and i <= (tone_index[2]+30)*15:
    tone_array[i] = 1
  else:
    tone_array[i] = 0
```

```python
#create predictors variable for extinction 
predictors_extinction = df(data={'Intercept': np.ones_like(tone_array), 'tone onset': tone_array})
modelB = sm.GLM(second_trace, predictors_extinction, family=Poisson())
modelB_results = modelB.fit() # Fit model to our data

# creating the r_index_extinction variable using np.zeros)
r_index_extinction = np.zeros(63)

for i in range (63):
  trace_temp = extinction_traces[:, i]
  modelB = sm.GLM(trace_temp, predictors_extinction, family=Poisson())
  modelB_results = modelB.fit() # Fit model to our data
  predicted_trace = modelB_results.predict()
  rB=pearsonr(trace_temp, predicted_trace)
  r_index_extinction[i] = rB.statistic
```

```python
#use pearsonr function to correlate indexs 
pearsonr(r_model_conditioning, r_index_extinction) 
```

```python
#create scatter plot r_index_conditioning and r_index_extinction 
plt.scatter(r_model_conditioning, r_index_extinction)
plt.xlabel('Conditioning Traces') #plot x label
plt.ylabel('Extinction Traces') #plot y label
```
