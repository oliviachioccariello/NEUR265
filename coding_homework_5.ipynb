```python
import matplotlib.pyplot as plt   #import matplot
import numpy as np #import numpy 
```

```python
url1 = 'https://raw.githubusercontent.com/oliviachioccariello/NEUR265/main/acc_hits_hw5.csv' #import acc_hits file
url2 = 'https://raw.githubusercontent.com/oliviachioccariello/NEUR265/main/acc_fa_hw5.csv' #import acc_fa file

acc_hits = np.loadtxt(url1, delimiter = ',') #create acc_hits data variable
acc_fa = np.loadtxt(url2, delimiter = ',') #create acc_fa data variable
```

```python
print(np.shape(acc_hits)) #print shape of acc_hits data
print(np.shape(acc_fa)) #print shape of acc_fa data
```

```python
time=np.linspace(0,4, num=8001) #create time variable 
```

```python
hz = 8001/4 #determine sampling rate 
print(hz)
```

```python
fig, axes = plt.subplots(2, 1)
axes[0].plot(time, acc_hits[0]) #plot first trial of acc_hits 
axes[1].plot(time, acc_fa[0]) #plot first trial of acc_fa
axes[0].set_ylabel('Hits')
axes[1].set_ylabel('False Alarms')
axes[1].set_xlabel('Time')
```
